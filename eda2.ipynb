```python
q1 = df.describe()["total_bill"]["25%"]
q3 = df.describe()["total_bill"]["75%"]
iqr = q3 - q1

lower_limit = q1 - 1.5 * iqr
upper_limit = q3 + 1.5 * iqr
```

```python
import tensorflow as tf
import numpy as np

# Assuming x_train and y_train are already defined and preprocessed

# Print shapes
print("x_train shape:", x_train.shape)  # Should print (195, 6)
print("y_train shape:", y_train.shape)  # Should print (195, 1)

# Reshape y_train if necessary (already correct in your case)
if y_train.ndim == 1:
    y_train = np.array(y_train).reshape(-1, 1)

# Define the model using an Input layer
demo = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(6,)),  # Specify the input shape directly
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')  # Output layer for binary classification
])

# Compile the model
demo.compile(optimizer=tf.keras.optimizers.Adam(0.001), 
             loss='binary_crossentropy', 
             metrics=['accuracy'])

# Fit the model
history = demo.fit(x_train, y_train, epochs=100, batch_size=32, validation_split=0.2)
```

```python
import matplotlib.pyplot as plt

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper right')
plt.show()

# Plot training & validation accuracy values
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```
