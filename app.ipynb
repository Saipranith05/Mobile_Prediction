```python
import pandas as pd
import zipfile
import os

# Path to the zip file
zip_file_path = r"C:\Users\Akshitha\New folder\Meachine_learning_pratice\archive (3).zip"

# Unzip the file to a temporary location
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall("temp_folder")

# Filter to find the CSV file
extracted_files = [f for f in os.listdir("temp_folder") if f.endswith('.csv')]
print("Extracted files:", extracted_files)

# Proceed only if there's at least one CSV file
if extracted_files:
    # Load the CSV file into a DataFrame
    csv_file_path = os.path.join("temp_folder", extracted_files[0])
    df = pd.read_csv(csv_file_path)

    # Rename columns
    df.columns = [
        "user_id", "device_model", "operating_system", "app_usage_time",
        "screen_on_time", "battery_drain", "number_of_apps_installed",
        "data_usage", "age", "gender", "user_behavior_class"
    ]

    # Display the DataFrame (optional)
    print(df.head())
else:
    print("No CSV file found in the extracted contents.")
```

```python
d1 = {"Xiaomi Mi 11": 0, "iPhone 12":1, "Google Pixel 5":2, "OnePlus 9":3, "Samsung Galaxy S21":4}
df["device_model"] = df["device_model"].map(d1)
```

```python
d2 = {"Android":0, "iOS":1}
df["operating_system"] = df["operating_system"].map(d2)
```

```python
d3 = {"Male":0, "Female":1}
df["gender"] = df["gender"].map(d3)
```

```python
df.drop(columns=["user_behavior_class"])
```

```python
x = df.drop(columns="gender")
y = df["gender"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```
